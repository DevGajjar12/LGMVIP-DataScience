***cv2.imread()*** method loads an image from the specified file. If the image cannot be read (because of missing file, improper permissions, unsupported or invalid format) then this method returns an empty matrix.

```python
import cv2 # for main project
import matplotlib.pyplot as plt  # for visualisation
```

# **Read the Image in RGB Format:**

```python
#Reading image
img = cv2.imread("Dhoni.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Converting BGR to RGB

# Displaying the original img
plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

# **Convert the RGB Image to Grayscale Image:**

```python
# We are considering gs_img as the variable for the GRAYSCALE IMAGE
gs_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8, 8))
plt.imshow(gs_img,cmap='gray')
plt.axis("off")
plt.title("Grayscale Image")
plt.show()
```

# **Converting Grayscale Image to INVERTED Grayscale:**

```python
# igs_img as the variable for the INVERTED Grayscale Image
igs_img = cv2.bitwise_not(gs_img)
plt.figure(figsize=(8, 8))
plt.imshow(igs_img, cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

# Smoothing the image with GaussianBlur Function:

OpenCV provides ***cv2.GaussianBlur()*** function to apply Gaussian Smoothing on the input source image. Following is the syntax of GaussianBlur() function :

dst = cv2.GaussianBlur(src, ksize, sigmaX[, dst[, sigmaY[, borderType=BORDER_DEFAULT]]] )


* Image blurring is achieved by convolving the image with a low-pass filter kernel. It is useful for removing noise. 
* It actually removes high frequency content (e.g: noise, edges) from the image resulting in edges being blurred when this is filter is applied. (Well, there are blurring techniques which do not blur edges). 
* OpenCV provides mainly four types of blurring techniques.:
    1. Averaging
    2. Gaussian Filtering
    3. Median Filtering
    4. Bilateral Filtering

For our project, we will be using the Second technique, Gaussian Filtering, using the **cv2.GaussianBlur()** function.
* We will have to specify the width and height of the kernel which should be positive and odd. 
* We also have to specify the standard deviation in the X and Y directions, sigmaX and sigmaY respectively. 
* If only sigmaX is specified, sigmaY is taken as equal to sigmaX. If both are given as zeros, they are calculated from the kernel size. 
* Gaussian filtering is highly effective in removing Gaussian noise from the image.


```python
# We are considering smooth_img as the variable for the SMOOTHENED IMAGE
smooth_img = cv2.GaussianBlur(igs_img, (29, 29), sigmaX=0, sigmaY=0)
plt.figure(figsize=(8, 8))
plt.imshow(smooth_img, cmap="gray")
plt.axis("off")
plt.title("Smoothened Image")
plt.show()
```

```python
# We are considering ps_img as the variable for the FINAL PENCIL SKETCH IMAGE
ps_img = cv2.divide(gs_img, 255 - smooth_img, scale=200)
plt.figure(figsize=(8, 8), linewidth=6)
plt.imshow(ps_img, cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
